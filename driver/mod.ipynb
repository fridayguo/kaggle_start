```python
import numpy as np
import pandas as pd 
import xgboost as xgb
import matplotlib.pylab as ply
```

```python
train = pd.read_csv('./driver/train.csv')
test = pd.read_csv('./driver/test.csv')

print 'train.shape:', train.shape
print 'test.shape:', test.shape
```

```python
X_train = train.drop(['id','target'], axis=1)
y_train = train['target']
X_test = test.drop(['id'], axis=1)
#y_test = test['target']
#X_train.info()
```

```python
from sklearn.model_selection  import cross_val_score
mod = xgb.XGBClassifier(max_depth=4, learning_rate=0.1, min_child_weight=5, 
                               n_estimators=100, objective='binary:logistic', seed=819 )
mod.fit(X_train, y_train)
mod
```

```python
y_pred = mod.predict_proba(X_test)
```

```python
output = pd.DataFrame({'id': test['id'].astype(np.int32), 'target': y_pred[:,1]})
output.to_csv('driver_xgb2.csv', index=False)
```

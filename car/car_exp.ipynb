```python
dtype_df = train_df.dtypes.reset_index()
dtype_df.columns = ['Feat','ColType']
dtype_df.groupby('ColType').count().reset_index()
```

```python
unique_values_dict = {}
for col in train_df.columns:
    value_count = len(train_df[col].unique())
    unique_values_dict.setdefault(value_count, [])
    unique_values_dict[value_count].append(col)
    
    #if train_df.dtypes[col] == 'object':
    #    print col
    
unique_values_dict
```
